```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
afgri=pd.read_csv('./Data/real_data.csv')
```

```python
afgri.time=afgri.time.str.replace('동영상 채팅','')
```

```python
afgri
```

```python
for i in range(0,len(afgri.time)):
    if len(afgri.time[i])==4:
        afgri.time[i]='0:0'+afgri.time[i]
    elif len(afgri.time[i])==5:
        afgri.time[i]='0:'+afgri.time[i]
```

```python
afgri=afgri.sort_values('time').reset_index().iloc[:,1:]
```

```python
afgri.time=pd.to_timedelta(afgri.time)
```

```python
afgri['time2seconds']=afgri.time.dt.total_seconds()
```
