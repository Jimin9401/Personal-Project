```python
dataset=pd.DataFrame(lis2,columns=['time','ID'])
```

```python
dataset=dataset[~dataset.time.str.contains('게시')]
dataset=dataset[~dataset.ID.str.contains('싹둑')]
```

```python
b=list(dataset.ID.str.split(':',n=1))
```
