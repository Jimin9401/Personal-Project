```python
from konlpy.tag import Twitter

tknz = Twitter()
pos_list_init = ["Noun", "Verb", "Exclamation" ,"KoreanParticle"]
pos_list_final = {"Noun":0, "Verb":0, "Exclamation":0 ,"KoreanParticle":0}

def tokenize(sentence):
    res = []
    tokens = tknz.pos(sentence)
    if len(tokens) <= 3:
        for token in tokens:
            if token[1] in pos_list_init:
                word = token[0]+"/"+token[1]
                res.append(word)
            else:
                try:
                    pos_list_final[token[1]]+=1
                except:
                    pos_list_final[token[1]]=1
                    print(token[1]+" is added to pos_list_final")
                
                word = token[0]+"/"+token[1]
                res.append(word)

    else:
        for token in tokens:
            if token[1] in pos_list_init:
                word = token[0]+"/"+token[1]
                res.append(word)
                pos_list_final[token[1]] +=1

    return res
tokenize(comment[0])
```

```python
token_comment_highlight=[tokenize(i) for i in comment_highlight]
token_comment_normal=[tokenize(i) for i in comment_normal]
```

```python
model_nor = Word2Vec(token_comment_normal, size=50, window=5, min_count=3, sg=1, workers=4, iter=500)
model_highlight=Word2Vec(token_comment_highlight, size=50, window=5, min_count=3, sg=1, workers=4, iter=500)
#model.save("word2vec_50_total.model")
```

```python
flat_nor=sum(token_comment_normal,[])
flat_highlight=sum(token_comment_highlight,[])
```

```python
dic_normal={}
for i in flat_nor:
    if i in dic_normal:
        dic_normal[i]+=1
    else:
        dic_normal[i]=1
```

```python
dic_hi={}
for i in flat_highlight:
    if i in dic_hi:
        dic_hi[i]+=1
    else:
        dic_hi[i]=1
```

```python
num_samples = 300
word_list = random.sample(list(model_nor.wv.vocab.keys()), num_samples)
embeddings = [model_nor[word] for word in word_list]

pca = PCA(n_components=2).fit_transform(embeddings)

plt.figure(figsize=(20, 20))
plt.scatter(pca[:,0],pca[:,1], c="r")
for i, txt in enumerate(word_list):
    text = txt.split("/")
    if len(text[0]) > 1:
        plt.annotate(text[0], (pca[i,0], pca[i,1]), size=22)
#plt.xlim(-0.75,0.75)
plt.show()
```

```python
word2ix = {"<PAD>":0, "<UNK>":1}
for i, v in enumerate(vocab):
    word2ix[v] = i+2
word2ix
```

```python
ix2word = {word2ix[k]:k for k in word2ix.keys()}
ix2word
```

```python
comment_ix = []

for token in token_comment:
    if len(token) != 0:
        seq = []
        for word in token:
            seq.append(word2ix[word])
    else:
        seq = []
        seq.append(word2ix["<UNK>"])
    comment_ix.append(seq)
```
